### How frequently does your first name occur across the years in the US among baby names?

```python
import pandas as pd
```

```python
us_baby_names = pd.read_csv("us_baby_names.csv")
```

```python
us_arf = us_baby_names.loc[us_baby_names["Name"] == "Arif", :]
```

```python
us_arf
```

```python
us_arf.plot.barh(x = "Year", y = "Count")
```
