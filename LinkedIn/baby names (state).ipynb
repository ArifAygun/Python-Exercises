## What were the 10 most popular baby names in 2000 in the State Texas?

```python
import pandas as pd
```

```python
state_baby_names = pd.read_csv ('state_baby_names.csv')
```

```python
state_baby_names
```

```python
state_baby_names_2000 = state_baby_names.loc[state_baby_names['Year'] == 2000, :]
```

```python
state_baby_names_2000
```

```python
tx_state_baby_names_2000 = state_baby_names_2000.loc[state_baby_names_2000['State'] == 'TX', :]
```

```python
tx_state_baby_names_2000
```

```python
sorted_tx_2000 = tx_state_baby_names_2000.sort_values('Count', ascending = False)
```

```python
sorted_tx_2000
```
